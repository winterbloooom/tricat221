```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
file_name = "/home/lumos/tricat/src/tricat221/src/2022-05-14_22-46-59.csv"
df = pd.read_csv(file_name, header=None)
```

```python
df
```

```python
raw_x = df.iloc[:, 2]
f_x = df.iloc[:, 5]

raw_y = df.iloc[:, 3]
f_y = df.iloc[:, 6]

raw_z = df.iloc[:, 4]
f_z = df.iloc[:, 7]

plt.figure(figsize=(25, 18))

plt.subplot(3, 1, 1)
plt.title('x')
plt.plot(raw_x, '.', color='dodgerblue', label='original')
plt.plot(f_x, 'r-', label='filtered_x')
plt.legend()

plt.subplot(3, 1, 2)
plt.title('y')
plt.plot(raw_y, '.', color='dodgerblue', label='original')
plt.plot(f_y, 'r-', label='filtered_y')
plt.legend()

plt.subplot(3, 1, 3)
plt.title('z')
plt.plot(raw_z, 'b-', color='dodgerblue', label='original')
plt.plot(f_z, 'r-', label='filtered_z')
plt.legend()

plt.tight_layout()
plt.show()
```
